$\textbf{Question 7}$

Consider the softplus activation function $𝑓(𝑥) = ln (1 + 𝑒^𝑥)$ utilised in machine learning. 
Using sagemath

(i) Compute the Taylor approximation of 𝑓(𝑥) around 𝑥 = 0 including terms up to 10-th order in 𝑥 (use series function). 

$\textbf{Answer}$

```python
# Define the softplus function
var('x')
softplus(x) = ln(1 + exp(x))

# Compute the Taylor series expansion around x = 0 up to the 10th order
taylor_approximation = softplus.series(x == 0, 10).simplify()

# Print the result
print(taylor_approximation)
```

(ii) Plot the softplus function as well as the Taylor approximation of question (i) in a combined
plot in the region −5 < 𝑥 < 5, −1 < 𝑦 < 4 (use different colors and styles for the two curves)
and comment on the region of validity of the approximation.

$\textbf{Answer}$

```python
# Plotting
p = plot(softplus, -5, 5, ymin=-1, ymax=4, color='blue', legend_label='Softplus Function')
p += plot(taylor_approximation, -5, 5, color='red', linestyle='--', legend_label='Taylor Approximation (10th order)')
p.show()
```

(iii) Compute the derivative of 𝑓(𝑥) analytically

$\textbf{Answer}$

```python
# Compute the derivative of the softplus function
derivative_softplus = diff(softplus, x)

# Print the result
print(derivative_softplus)
```

(iv) Compute the derivative of 𝑓(𝑥) approximately using the result of (i).

$\textbf{Answer}$

```python
# Differentiate the Taylor approximation term by term
derivative_taylor = taylor_approximation.diff(x)

# Print the result
print(derivative_taylor)
```

(v) Plot the results of (iii) and (iv) in a combined plot in the region −5 < 𝑥 < 5, −1 < 𝑦 < 1
(use different colors and styles for the two curves).

$\textbf{Answer}$

```python
# Plotting
p = plot(derivative_softplus, -5, 5, ymin=-1, ymax=1, color='blue', legend_label='Analytical Derivative')
p += plot(derivative_taylor, -5, 5, color='yellow', linestyle='--', legend_label='Approximate Derivative (Taylor)')
p.show() 
```
