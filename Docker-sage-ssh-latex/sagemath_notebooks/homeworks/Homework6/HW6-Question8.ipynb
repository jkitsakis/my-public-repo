$\textbf{Question8}$  

Given the function $f(x_1, x_2) = e^{x_1}(4x_1^{2} + 2x_2^{2} + 4x_1x_2 + 2x_2 + 1)$ and working with sagemath:

(a) Find analytically and classify (e.g. minimum,maximum,saddle point) the stationary points(at finite $x_1$, $x_2$).

(b) Plot the above function using a contour plot (with appropriate parameter bounds) and depict the position of the stationary points. Use axes labels, contour labels and an appropriate number of contours.

(c) Working numerically, find the global minimum (at finite $x_1$, $x_2$) starting from the point $x_0 = (âˆ’1, 2)$ and utilizing two different algorithms (one using the derivatives and one not using them). Comment on the results compared to the results of (a).

$\textbf{Answer}$ 

```python
# Define the function
f(x1, x2) = exp(x1)*(4*x1^2 + 2*x2^2 + 4*x1*x2 + 2*x2 + 1)

# (a) Find stationary points analytically
var('x1 x2')
grad_f = f.gradient()(x1, x2)
stationary_points = solve([eq == 0 for eq in grad_f], x1, x2)
stationary_points
```

```python
# Define the function
f(x1, x2) = exp(x1)*(4*x1^2 + 2*x2^2 + 4*x1*x2 + 2*x2 + 1)

# Plot the function with contour plot
contour = contour_plot(f, (x1, -5, 5), (x2, -5, 5), contours=20, fill=False, cmap='cool', labels=True, label_inline=True, axes_labels=['$x_1$', '$x_2$'])


# Find and plot stationary points
stationary_points = solve([eq == 0 for eq in grad_f], x1, x2)
for p in stationary_points:
     contour += point((p[0].rhs(), p[1].rhs()), color='red', size=30)

# Show the plot
show(contour)
```
