$\textbf{Q1.}$ For x  a real number, calculate the determinant of the matrix $C=\begin{bmatrix} cosx & 0 & sinx \\ 0 & 1 & 0 \\ -sinx & 0 & cosx \end{bmatrix}$ using Laplace Expansion
\
\
$\textbf{Answer :}$


To calculate the determinant of the matrix $C$, denoted as $\text{det}(C)$, using Laplace expansion, we can expand along any row or column. 
\
Let's choose the first column for this example.
\
\
The determinant is given by:
\
$
\text{det}(C) = \sum_{i=1}^{n} (-1)^{i+j} \cdot a_{ij} \cdot \text{det}(M_{ij})
$
\
\
where $a_{ij}$ is the element in the $i$-th row and $j$-th column of the matrix, and $M_{ij}$ is the matrix obtained by removing the $i$-th row and $j$-th column from the original matrix.
\
\
For our matrix $C$:
\
$
C = \begin{bmatrix}
\cos(x) & 0 & \sin(x) \\
0 & 1 & 0 \\
-\sin(x) & 0 & \cos(x)
\end{bmatrix}
$
\
\
Let's expand along the first column:
\
\
$
\text{det}(C) = \cos(x) \cdot \text{det}\begin{bmatrix} 1 & 0 \\ 0 & \cos(x) \end{bmatrix} - 0 \cdot \text{det}\begin{bmatrix} 0 & \sin(x) \\ 0 & \cos(x) \end{bmatrix} + \sin(x) \cdot \text{det}\begin{bmatrix} 0 & 1 \\ -\sin(x) & 0 \end{bmatrix}
$
\
\
Now, calculate the determinants of the smaller matrices:
\

$
\text{det}\begin{bmatrix} 1 & 0 \\ 0 & \cos(x) \end{bmatrix} = 1 \cdot \cos(x) - 0 \cdot 0 = \cos(x)
$
\
\
$
\text{det}\begin{bmatrix} 0 & \sin(x) \\ 0 & \cos(x) \end{bmatrix} = 0 \cdot \cos(x) - \sin(x) \cdot 0 = 0
$
\
\
$
\text{det}\begin{bmatrix} 0 & 1 \\ -\sin(x) & 0 \end{bmatrix} = 0 \cdot 0 - 1 \cdot (-\sin(x)) = \sin(x)
$
\
\
Now substitute these back into the Laplace expansion formula:
\
\
$
\text{det}(C) = \cos(x) \cdot \cos(x) - 0 + \sin(x) \cdot \sin(x) = \cos^2(x) + \sin^2(x)
$
\
\
Using the trigonometric identity $\cos^2(x) + \sin^2(x) = 1$, we find that:
\
\
$
\text{det}(C) = 1
$
\
\
So, $\text{det}(C) = 1$ for any real number $x$.
