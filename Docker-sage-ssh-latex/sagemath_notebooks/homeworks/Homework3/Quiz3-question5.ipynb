$\textbf{Q5.}$ Let $A∈R^{2×2}$  such that $det(A)=3$. Calculate $det(3A^TA^{−1})$ 
\
\
$\textbf{Answer :}$


Let $A \in \mathbb{R}^{2 \times 2}$ such that $\det(A) = 3$. Calculate $\det(3A^TA^{-1})$.
\
\
Using the property that $\det(cB) = c^n \det(B)$, where $c$ is a scalar (In our case, c=3 and n=2) 
\
and $B$ is an $n \times n$ matrix, we have:
\
\
$
\det(3A^T) = 3^2 \det(A^T)
$
\
\
Since $\det(A^T) = \det(A)$, we substitute:
\
\
$
\det(3A^T) = 3^2 \det(A)
$
\
\
Given that $\det(A) = 3$, we substitute this value in:
\
\
$
\det(3A^T) = 3^2 \cdot 3
$
\
\
Calculating the result:
\
\
$
\det(3A^T) = 27
$
\
\
Next, we need to calculate $\det(A^{-1})$. The determinant of the inverse of a matrix is the reciprocal of the determinant of the original matrix:
\
\
$
\det(A^{-1}) = \frac{1}{\det(A)}
$
\
\
Substitute $\det(A) = 3$:
\
\
$
\det(A^{-1}) = \frac{1}{3}
$
\
\
Now, multiply the determinants:
\
\
$
\det(3A^TA^{-1}) = \det(3A^T) \cdot \det(A^{-1})
$
\
\
$
\det(3A^TA^{-1}) = 27 \cdot \frac{1}{3}
$
\
\
$
\det(3A^TA^{-1}) = 9
$
\
\
Therefore, $\det(3A^TA^{-1}) = 9$.
