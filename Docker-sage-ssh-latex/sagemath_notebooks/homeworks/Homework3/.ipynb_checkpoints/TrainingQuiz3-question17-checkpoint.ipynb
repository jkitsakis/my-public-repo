$\textbf{Q17.} $Determine the algebraic and geometric multiplicity of each eigenvalue of the matrix 
$A=\begin{bmatrix} 1 & 0 & 2 \\ -1 & 1 & 3 \\ 0 & 0 & 2 \end{bmatrix}$
\
\
$\textbf{Answer :}$


To find the algebraic and geometric multiplicities of each eigenvalue of a matrix, we first need to find the eigenvalues. The eigenvalues are the solutions to the characteristic equation, which is obtained by setting the determinant of the matrix minus $\lambda$ times the identity matrix equal to zero:

$
\text{det}(A - \lambda I) = 0
$

For the given matrix $A$:

$
A = \begin{bmatrix} 1 & 0 & 2 \\ -1 & 1 & 3 \\ 0 & 0 & 2 \end{bmatrix}
$

Let's find the eigenvalues:

$
\text{det}(A - \lambda I) = \begin{vmatrix} 1 - \lambda & 0 & 2 \\ -1 & 1 - \lambda & 3 \\ 0 & 0 & 2 - \lambda \end{vmatrix}
$

Expanding the determinant:

$
(1 - \lambda)((1 - \lambda)(2 - \lambda) - (0)(0)) - (0 - (-1)(2))(0 - (0)(3))
$

$
= (1 - \lambda)(2 - \lambda) + 2 = 0
$

Solving for $\lambda$, we get two eigenvalues:

$
\lambda_1 = 1
$

$
\lambda_2 = 2
$

Now, let's find the corresponding eigenvectors. For each eigenvalue $\lambda$, solve the system of linear equations $(A - \lambda I)\mathbf{v} = \mathbf{0}$ to find the eigenvector $\mathbf{v}$.

For $\lambda = 1$:

$
(A - \lambda I)\mathbf{v} = \begin{bmatrix} 0 & 0 & 2 \\ -1 & 0 & 3 \\ 0 & 0 & 1 \end{bmatrix}\mathbf{v} = \mathbf{0}
$

Solving this system gives the eigenvector:

$
\mathbf{v}_1 = \begin{bmatrix} 2 \\ -3 \\ 0 \end{bmatrix}
$

For $\lambda = 2$:

$
(A - \lambda I)\mathbf{v} = \begin{bmatrix} -1 & 0 & 2 \\ -1 & -1 & 3 \\ 0 & 0 & 0 \end{bmatrix}\mathbf{v} = \mathbf{0}
$

Solving this system gives the eigenvector:

$
\mathbf{v}_2 = \begin{bmatrix} -2 \\ -3 \\ 1 \end{bmatrix}
$

Now, let's determine the algebraic and geometric multiplicities for each eigenvalue:

For $\lambda_1 = 1$:
- Algebraic multiplicity ($m_a(\lambda_1)$) is the number of times $\lambda_1$ appears as a root of the characteristic equation, which is 1.
- Geometric multiplicity ($m_g(\lambda_1)$) is the dimension of the eigenspace corresponding to $\lambda_1$, which is 1.


For $\lambda_2 = 2$:

- Algebraic multiplicity ($m_a(\lambda_2)$) is the number of times $\lambda_2$ appears as a root of the characteristic equation, which is 2.
- Geometric multiplicity ($m_g(\lambda_2)$) is the dimension of the eigenspace corresponding to $\lambda_2$, which is 1.


So, the algebraic and geometric multiplicities for each eigenvalue are:

$
\lambda_1 = 1: \quad m_a(\lambda_1) = 1, \quad m_g(\lambda_1) = 1
$

$
\lambda_2 = 2: \quad m_a(\lambda_2) = 2, \quad m_g(\lambda_2) = 1
$
