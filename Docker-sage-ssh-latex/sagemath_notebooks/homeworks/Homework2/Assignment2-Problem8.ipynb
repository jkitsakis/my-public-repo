\
Given a non-zero vector $ v ∈ R^m$ the Householder matrix $ P$  is defined as
$ P = 1 − βuu^T , β = 2/(u^T u)$  where 1 is the m-by-m identity matrix. 
\
If a vector $ x ∈ R^m $ is multiplied by $ P$ , then it is reflected in the hyperplane $ {span(v)}^⊥$ . 
\
Given $  v = (1, −1, 1)^T $ use Sagemath to 
\
(a) compute the associated Householder matrix P as well as $ P^ 2 $
\
\
$\textbf{Answer :}$

```python
# Given vector v
v = vector([1, -1, 1])

# Compute b
b = 2 / v.norm()^2

# Compute uu^T
uut = v.column() * v.row()
print(v.column() , v.row())
print("uut:")
print(uut)

# Compute Householder matrix P
P = identity_matrix(3) - b * uut

# Compute P^2
P_squared = P * P

# Display the results
print("Householder Matrix P:")
print(P)
print("\nHouseholder Matrix P^2:")
print(P_squared)
```

(d) plot the vectors v, x, y in a common 3-dimensional plot.
\
\
$\textbf{Answer :}$

```python
# Given vectors v, x, and y
v = vector([1, -1, 1])
x = vector([1, 1, 3])
y = vector([-1, 3, 1])

# Create a 3D plot with arrows for vectors
p = arrow3d((0, 0, 0), v, color='red', legend_label='v')
p += arrow3d((0, 0, 0), x, color='blue', legend_label='x')
p += arrow3d((0, 0, 0), y, color='green', legend_label='y')

# Display the plot
show(p, aspect_ratio=(1, 1, 1))
```
