$\textbf{Problem 10}$
\
\
Consider the vector space R^4 with a Euclidean inner product. You are given three vectors 
$(1, 0, 1, 0)^T , (2, 2, 0, −3)^T , (0, 3, 6, 5)^T$ .
\
Using by-hand calculation :
\
(a) verify that these vectors are linearly independent;
\
\
$\textbf{Answer :}$

To check if the vectors that span $U$ are linearly independent, we need to check if there exist constants $c_1v_1 + c_2v_2 + c_3v_3 =0 $ such that $c_1=c_2=c_3=0$ 
\
\
The vectors are:
$ 
v_1 =  \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix}
v_2 =  \begin{bmatrix} 2 \\ 2 \\ 0 \\ 3 \end{bmatrix}
v_3 =  \begin{bmatrix} 0 \\ 3 \\ 6 \\ 5 \end{bmatrix}
$
\
So it is
\
$c_1v_1 + c_2v_2 + c_3v_3 =0 => 
c_1 \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix} + c_2\begin{bmatrix} 2 \\ 2 \\ 0 \\ 3 \end{bmatrix} + c_3\begin{bmatrix} 0 \\ 3 \\ 6 \\ 5 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 0 \\ 0 \end{bmatrix} => $
\

$c_1 + 2c_2 + 0c_3 = 0 $
\
$0c_1 + 2c_2 + 3c_3 = 0 $
\
$c_1 + 0c_2 + 6c_3 = 0 $
\
$0c_1 -3c_2 + 5c_3 = 0 $
\
\
So the coefficient matrix is 
$A =  \begin{bmatrix} 1 & 2 & 0 \\ 0 & 2 & 3 \\ 1 & 0 & 6 \\ 0 & -3 & 5  \end{bmatrix} $ 
\
\
$R_3 = R_3 - R_1 -> \begin{bmatrix} 1 & 2 & 0 \\ 0 & 2 & 3 \\ 0 & -2 & 6 \\ 0 & -3 & 5  \end{bmatrix} $ 
\
\
$R_2 = 1/2 R_2  -> \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 3/2 \\ 0 & -2 & 6 \\ 0 & -3 & 5  \end{bmatrix} $ 
\
\
$R_3 = R_3 - 2R_2 -> \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 3/2 \\ 0 & 0 & 3 \\ 0 & -3 & 5  \end{bmatrix} $
\
\
$R_4 = R_4 - 3R_2 -> \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 3/2 \\ 0 & 0 & 3 \\ 0 & 0 & -1/2  \end{bmatrix} $
\
\
$R_3 = 1/3 R_3 -> \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 3/2 \\ 0 & 0 & 1 \\ 0 & 0 & -1/2  \end{bmatrix} $
\
\
$R_1 = R_1 - 2R_3 -> \begin{bmatrix} 1 & 0 & -3 \\ 0 & 1 & 3/2 \\ 0 & 0 & 1 \\ 0 & 0 & -1/2  \end{bmatrix} $
\
\
$R_4 = R_4 + 1/2R_3 -> \begin{bmatrix} 1 & 0 & -3 \\ 0 & 1 & 3/2 \\ 0 & 0 & 1 \\ 0 & 0 & 0  \end{bmatrix} $

The system is consistent and has a unique solution, indicating that the given vectors are linearly independent.


\
\
$\textbf{(b) Apply the Gram–Schmidt procedure to transform the above basis vectors to obtain an orthonormal
basis.}$
\
\
$\textbf{Answer :}$


Given vectors:
$
\mathbf{v}_1 = \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix}, \quad
\mathbf{v}_2 = \begin{bmatrix} 2 \\ 2 \\ 0 \\ -3 \end{bmatrix}, \quad
\mathbf{v}_3 = \begin{bmatrix} 0 \\ 3 \\ 6 \\ 5 \end{bmatrix}.
$

$\textbf{Gram-Schmidt Process:}$

$\textbf{1. Orthogonalization:}$

$
\mathbf{u}_1 = \mathbf{v}_1 = \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix},
$

$
\mathbf{u}_2 = \mathbf{v}_2 - \text{proj}_{\mathbf{u}_1}(\mathbf{v}_2) = \begin{bmatrix} 1 \\ 2 \\ -1 \\ -3 \end{bmatrix},
$

$
\mathbf{u}_3 = \mathbf{v}_3 - \text{proj}_{\mathbf{u}_1}(\mathbf{v}_3) - \text{proj}_{\mathbf{u}_2}(\mathbf{v}_3) = \begin{bmatrix} -2 \\ 5 \\ 2 \\ 2 \end{bmatrix}.
$

$\textbf{Projection Calculations:}$

$
\text{proj}_{\mathbf{u}_1}(\mathbf{v}_2) = \frac{\mathbf{v}_2 \cdot \mathbf{u}_1}{\|\mathbf{u}_1\|^2} \cdot \mathbf{u}_1 = \frac{2}{2} \cdot \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix},
$

$
\text{proj}_{\mathbf{u}_2}(\mathbf{v}_3) = \frac{\mathbf{v}_3 \cdot \mathbf{u}_2}{\|\mathbf{u}_2\|^2} \cdot \mathbf{u}_2 = \frac{\begin{bmatrix} 0 \\ 3 \\ 6 \\ 5 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 2 \\ -1 \\ -3 \end{bmatrix}}{\|\begin{bmatrix} 1 \\ 2 \\ -1 \\ -3 \end{bmatrix}\|^2} \cdot \begin{bmatrix} 1 \\ 2 \\ -1 \\ -3 \end{bmatrix} = \begin{bmatrix} -1 \\ -2 \\ 1 \\ 3 \end{bmatrix},
$

$
\text{proj}_{\mathbf{u}_1}(\mathbf{v}_3) = \frac{\mathbf{v}_3 \cdot \mathbf{u}_1}{\|\mathbf{u}_1\|^2} \cdot \mathbf{u}_1 = \frac{\begin{bmatrix} 0 \\ 3 \\ 6 \\ 5 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix}}{\|\begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix}\|^2} \cdot \begin{bmatrix} 1 \\ 0 \\ 1 \\ 0 \end{bmatrix} = \begin{bmatrix} 3 \\ 0 \\ 3 \\ 0 \end{bmatrix}.
$


$\textbf{2. Normalization:}$

$
\|\mathbf{u}_1\| = \sqrt{2}, \quad \|\mathbf{u}_2\| = \sqrt{15}, \quad \|\mathbf{u}_3\| = \sqrt{37}.
$

$
\mathbf{e}_1 = \frac{\mathbf{u}_1}{\|\mathbf{u}_1\|} = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ 0 \\ \frac{1}{\sqrt{2}} \\ 0 \end{bmatrix},
$

$
\mathbf{e}_2 = \frac{\mathbf{u}_2}{\|\mathbf{u}_2\|} = \begin{bmatrix} \frac{1}{\sqrt{15}} \\ \frac{2}{\sqrt{15}} \\ -\frac{1}{\sqrt{15}} \\ -\frac{3}{\sqrt{15}} \end{bmatrix},
$

$
\mathbf{e}_3 = \frac{\mathbf{u}_3}{\|\mathbf{u}_3\|} = \begin{bmatrix} -\frac{2}{\sqrt{37}} \\ \frac{5}{\sqrt{37}} \\ \frac{2}{\sqrt{37}} \\ \frac{2}{\sqrt{37}} \end{bmatrix}.
$


So, the orthonormal basis is given by $\{\mathbf{e}_1, \mathbf{e}_2, \mathbf{e}_3\}$.
