# 2. Solution of systems of equations

```python
x,y=var('x,y')
%display latex
```

```python
s=solve([7*x-2*y==1,2*x-3*y==4],[x,y])
s
```

```python
[s[0][0].rhs().n(digits=5),s[0][1].rhs().n(digits=5)]
```

Solution in domain

```python
solve(x^3-2*x^2+10*x==0,x)
```

```python
solve(x^3-2*x^2+10*x==0,x,algorithm='sympy',domain='real')
```

Substitution of solution(s)

```python
x,y=var('x,y')
a=x^2-y^2;
s=solve([7*x-2*y==1,2*x-3*y==4],[x,y]);
a.subs(s[0])
```

### Graphical solution

```python
x,y=var('x,y')
g1=implicit_plot(7*x-2*y==1,(x,-2,2),(y,-2,2),color='blue')
g2=implicit_plot(2*x-3*y==4,(x,-2,2),(y,-2,2),color='orange')
show(g1+g2,axes=True,frame=False,axes_labels=['$x$','$y$'],gridlines=True)
```
